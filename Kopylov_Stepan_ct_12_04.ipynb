**Librosa**

Я хочу представить библиотеку Librosa для работы с аудио в Python. Краткий пример пример кода:

```python
import librosa
filename = librosa.example('nutcracker')
y, sr = librosa.load(filename)
tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr)
print('Estimated tempo: {:.2f} beats per minute'.format(tempo))
beat_times = librosa.frames_to_time(beat_frames, sr=sr)
```

**Разберём каждую строку кода**

Во второй строке кода строковая прерменная filename принимает в себя путь к файлу примера.

В третьей строке в переменную y будет загружена звуковая волна, представляющая собой числовой массив, обонначающий амплетуду звука в разные моменты времени. В переменную sr будет загружена частота дискретизации аудио. Частота дискретизации определяет, сколько сэмплов (измерений звука) записывается в секунду.

Рассмотрим четвертую строку:
Здесь в переменнюу tempo будет сохранено значение темпа музыки (скорость ударов в минуту), а в переменную beat_frames - будут сохранены номера кадров, соответствующих обнаруженным событиям ритма.

Пятая строка выводит переменную tempo, то есть скорость ударов в минуту.

Следующая операция преобразует номера кадров beat_frames в тайминги:
теперь beat_timesэто будет массив временных меток (в секундах), соответствующих обнаруженным событиям ритма.



 **Разберем функции библиотеки, используемые в программе**

 Во второй строке использованна функция librosa.example, она возвращает путь к аудиофайлу.
 В третьей строке - load, она заргружает аудиофайл по указанному пути и возвращает временной ряд аудиосигнала и частоту дискретизации.
 В четвёртой - beat.beat_track, эта функция используется для определения темпа и индексов фреймов, соответствующих ударам.
 Последняя функция в этой программе - frames_to_time, она преобразует индексы фреймов ударов в соовтветствующие временные значения.
